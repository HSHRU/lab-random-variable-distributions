```python
import pandas as pd
import numpy as np

from scipy.stats import boxcox 

import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
ran_var_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
ran_var_df.columns= ran_var_df.columns.str.lower().str.replace(' ', '_')
```

```python
num = ran_var_df._get_numeric_data()
num.columns
```

```python
for col in num:
    sns.displot(num[col], bins=100)
```

```python
def log_transfom_clean1(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0
num_log = num.copy()

for col in num_log:
    num_log[col] = num_log[col].apply(lambda x :log_transfom_clean1(x))
```

```python
for col in num:
    sns.displot(num[col], bins=100)
    sns.displot(num_log[col], bins=100) 
```
